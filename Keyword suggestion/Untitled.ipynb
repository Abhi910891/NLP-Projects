```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import re 
from collections import Counter 
```

# File Opening and cleaning (change formate to utf-8)

```python
word = [] 

with open('autocorrect book.txt','r',encoding='utf-8') as f: 
    data = f.read() 
    data = data.lower() 
    words = re.findall('\w+',data) 
    words += words 
print(words[0:10])
```

```python
v = set(words) # convert it to sets
```

```python
words_freq_dict ={} 
words_freq_dict = Counter(words)
```

```python
words_freq_dict.most_common()[:10]
```

# Relative Frequency of words 

```python
words_freq_dict.values()
```

```python
Total = sum(words_freq_dict.values())
probs = {} 

for k in words_freq_dict.keys(): 
    probs[k] = words_freq_dict[k]/Total
```

```python
probs
```

# Finding the similar words

```python
import textdistance 
```

```python
def autocorrect(word):
    word = word.lower() 
    if word in v :
        return("Your word seems to be correct",word) 
    else:
        similarities = [1-(textdistance.Jaccard(qval=2).distance(v,word)) for v in words_freq_dict.keys()] 
        df = pd.DataFrame.from_dict(probs , orient="index").reset_index() 
        df = df.rename(columns={"index":"Word",0:"Probs"}) 
        df["Similarity"] = similarities 
        output = df.sort_values(["Similarity","Probs"],ascending=False).head(5)
        return output 
        
```
